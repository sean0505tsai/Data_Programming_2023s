````python
def evaluate(exp):
    print(exp + " =", eval(exp))
    
def newline(row=1):
    print( '\n' * (row-1) )

def markline(length=15, mark='-'):
    print(mark * length)
    
'''
int(string [, base])  # [內容] 方括號內的"內容"表示可選。
int(value) -> int
float(string) -> float
eval(expression) -> value: int or float, that depends on the type of result

type, id
'''

'''
先定義要用到的變數，
接著呼叫並將需要計算的式子，以字串的方式傳入 evalute(要計算的式子):
'''    
# list, tuple, dict, set
lst1 = [1, 2]
lst2 = [3, 4, 5]
t1 = (1, 2)
t2 = (3, 4, 5)

print('lst1:', lst1)
print('lst2:', lst2)
print('t1:', t1)
print('t2:', t2)
newline()

evaluate('lst1 + lst2')
evaluate('t1 + t2')
newline()

print("lst1 * 3 =", lst1 * 3)
print("t2 * 2 =", t2 * 2)
newline()

dct1 = {1:10, 2:40, 3:60}
dct2 = {1:1, 2:4, 3:6}
# evaluate('dct1 + dct2')
evaluate('1 + 2 + 3')
````
