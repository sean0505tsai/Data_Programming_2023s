#### Problem 1
- 請建立 Triangle 類別，需要給三個邊(a, b, c)的值。
- 同樣可以計算 area, perimeter, 與顯示(about)。
- 由於三角形的邊長需要滿足: 任兩邊和大於第三邊的規則。
- 所以，如果給的邊長不能構成三角形的話，就將其改成每邊為 1 的正三角形。
- 另外，area 需要使用 Heron's formula 來算。 

```python
import math

class Triangle:
    def __init__(self, a, b, c):
        if(a+b>c and a+c>b and b+c>a):
            self.a = a
            self.b = b
            self.c = c
        else:
            self.a = 1
            self.b = 1
            self.c = 1

    def get_perimeter(self):
        return self.a + self.b + self.c
    
    def get_area(self):
        s = (self.a + self.b + self.c)/2
        area = math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c))
        return area
    
    def about(self):
        print(f"Side A: {self.a}")
        print(f"Side B: {self.b}")
        print(f"Side C: {self.c}")
        print(f"Area: {self.get_area()}")
        print(f"Perimeter: {self.get_perimeter()}")

A = Triangle(3, 2, 1)
A.about()
```

#### Problem 4
- 讀入資料檔 BigData-Seat-2023-0302.txt，並完成下列的改寫:
- 姓名的部分: 是 2 個字的，將單名改成 *； 是 3 個字的中間那個字改成 *。
- 最後再將改寫過的結果，存入檔案 BigData-Seats.txt (encoding='utf-8')

```python
fname = './BigData-Seat-2023-0302.txt'
with open(fname, "r", encoding='utf-8') as fp:
    lines = fp.read()
    lines = lines.split('\n')

result = []
for line in lines:  # line = '108590099,1,7,王*同'
    items = line.split(',')
    #items[0] = items[0][:-2] + "**"
    if len(items[3]) > 2:
        items[3] = items[3][0] + '*' + items[3][2:]
    else:
        items[3] = items[3][0] + '*'
    
    line = ','.join(items)    
    result.append(line)
    
print(result)
with open('BigData-Seat.txt', "w", encoding='utf-8') as fp:
    fp.writelines('\n'.join(result))
```

```python
fname = './BigData-Seat.txt'
seat = [[] for i in range(4)]
result = "<table>\n"

for row in seat:
    for i in range(14):
        row.append(0)

with open(fname, "r", encoding='utf-8') as fp:
    lines = fp.read()
    lines = lines.split('\n')

for line in lines:
    # line = '108590099,1,7,王*同'
    items = line.split(',')
    row = int(items[1])
    col = int(items[2])
    # print(items[2])
    if(col < 0):
        seat[row-1][14+col] = f"{items[0]}<br>{items[3]}"
    else:
        seat[row-1][col-1] = f"{items[0]}<br>{items[3]}"

for row in seat:
    result += "\t<tr>\n"
    for col in row:
        result += f"\t\t<th>{col}</th>\n"
    result += "\t</tr>\n"
result += "</table>"

with open('BigData-Seat.html', "w", encoding='utf-8') as fp:
    fp.writelines(result)

```
