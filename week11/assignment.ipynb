## 題目：2
請使用 matplotlib.pyplot 來繪製下列三角函數的圖形。  
含 x-軸，共有五條曲線，定義如下:  
1. sin(x): red。
2. cos(2x): green。
3. 0.5*sin(3x): blue。
4. sin(x) + cos(x): mangenta。
5. x-軸: gray。  

角度範圍: 0 ~ 360 or 0 ~ 2*pi，請依指定顏色畫出每個函式的圖形。

<img src='ShowingCurves.jpg'>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)         # sin(x)
y2 = 0.5*np.cos(2*x)
y3 = np.sin(3*x)
y4 = np.sin(x) + np.cos(x)
y5 = 0*x

x = (180/np.pi)*x

plt.title('108820018蔡翔宇')
plt.ylim(-2.0, 2.0)   #設定y軸範圍
plt.plot(x, y1, color = 'red', label = 'sin(x)')     # sin
plt.plot(x, y2, color = 'green', label = '0.5*cos(2x)')    # 0.5*cos(2x)
plt.plot(x, y3, color = '#0000FF', label = 'sin(3x)')    # 0.5*cos(2x)
plt.plot(x, y4, color = 'pink', label = 'sin(x)+cos(x)')     # sin(x) + cos(x)
plt.plot(x, y5, color = '#888888', label = 'x-axis')  # y = 0
plt.legend()
plt.show()
```

## 題目：3
請完成柱狀圖疊加形式的顯示。  
底下是年度與城市人口數的資料，請使用柱狀圖(bar)先畫 城市A，再疊加 城市B 的人口資料。  
中文字型，請使用標楷體，設定方式可參考右列網站。 [關於字型的設定](https://www.796t.com/content/1545046029.html)  
year = [2018, 2019, 2020, 2021, 2022]  
cityA = [128, 150, 199, 180, 150] <- 填充顏色: tomato。  
cityB = [120, 145, 180, 170, 120] <- 以預設顏色來填充。  
標題，座標軸等的訊息，請依照圖形中的內容進行設定。  

<img src='BarStackShow.jpg'>

```python
# p.11
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager

year = [2018,2019,2020,2021,2022]
cityA = [128,150,199,180,150]
cityB = [120,145,180,170,120]

plt.rcParams["font.sans-serif"] = "DFKai-SB"        # 標楷體

plt.bar(
    year, 
    cityA, 
    width=0.5, 
    color='tomato', 
    label='城市A'
    )

plt.bar(
    year, 
    cityB, 
    width=0.5,
    bottom=cityA,
    label = '城市B' 
)

plt.legend()
plt.title("108820018 蔡翔宇")
plt.xlabel("人口統計年度")
plt.ylabel("人口數")
plt.show()
```
